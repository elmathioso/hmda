```python
import pandas as pd
import json
```

```python
# Create column format dictionary
coldf = pd.read_csv('C:/data/HMDA/HMDA_columns_1981_1989.csv')
cols = json.loads(coldf.to_json(orient='records'))
cols
```

```python
#datarange = [x for x in range(1990, 2001, 1)]
datarange = [x for x in range(1981, 1982, 1)]
datarange
```

```python
first_file = True
for i in datarange:
    print(i)
    folder = "c:/data/HMDA/"
    
    
    filename = folder + "HMD_FACDSB" + str(i)[-2:] + ".txt"
    print(filename)
    
#     with open(filename) as f:
#         line = f.readline()
#         print(line)
   
    if first_file == True:
        df = pd.read_csv(filename, header=None, delimiter="\t", names=["original"])
    else:
        df.append(pd.read_csv(filename, header=None, delimiter="\t", names=["original"]))
```

```python
df.info()
```

```python
for col in cols:
    print(col['NAME'])
    df[col['NAME']] = df.original.map(lambda x: x[col["START"]:col["STOP"]])
```
